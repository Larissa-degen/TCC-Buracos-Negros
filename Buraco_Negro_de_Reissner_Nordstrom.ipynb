```python
from sympy import *
from sympy.plotting.plot import plot_parametric
```

```python
def figura_Reissner_Nordstrom(M, Q):
    theta = Symbol('theta')
    r_hor_m = M - sqrt(M**2 - Q**2)
    r_hor_p = M + sqrt(M**2 - Q**2)
    p = plot_parametric((r_hor_m*sin(theta), r_hor_m*cos(theta)), (r_hor_p*sin(theta), r_hor_p*cos(theta)), 
                        (theta, 0, 2*pi), aspect_ratio=(1, 1), xlim=[-10, 10], ylim=[-10, 10], 
                        xlabel="x", ylabel="z", legend=True, 
                        title="Reissner-Nordström, M={}, Q={}\n".format(M, Q), size=(6, 6), show=False)
    p[0].line_color = 'red'
    p[0].label = 'hor-'
    p[1].line_color = 'blue'
    p[1].label = 'hor+'
    p.show()
```

```python
figura_Reissner_Nordstrom(4.2, 2.0)
```

```python
figura_Reissner_Nordstrom(4.2, 3.9)
```
