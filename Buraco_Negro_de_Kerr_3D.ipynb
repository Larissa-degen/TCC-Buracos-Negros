```python
from sympy import *
from sympy.plotting.plot import plot3d_parametric_surface
```

```python
def figura_Kerr_3D(M, a):
    theta, phi = symbols('theta, phi')
    r_erg_m = M - sqrt(M**2 - (a**2)*(cos(theta)**2))
    r_erg_p = M + sqrt(M**2 - (a**2)*(cos(theta)**2))
    r_hor_m = M - sqrt(M**2 - a**2)
    r_hor_p = M + sqrt(M**2 - a**2)
    p = plot3d_parametric_surface((sqrt(r_erg_m**2 + a**2)*cos(phi)*sin(theta), 
                                   sqrt(r_erg_m**2 + a**2)*sin(phi)*sin(theta), r_erg_m*cos(theta), 
                                   (phi, 0.0*pi, 2*pi), (theta, 0, pi)), 
                                  (sqrt(r_hor_m**2 + a**2)*cos(phi)*sin(theta), 
                                   sqrt(r_hor_m**2 + a**2)*sin(phi)*sin(theta), r_hor_m*cos(theta),
                                   (phi, 0.0*pi, 1.0*pi), (theta, 0, pi)), 
                                  (sqrt(r_hor_p**2 + a**2)*cos(phi)*sin(theta), 
                                   sqrt(r_hor_p**2 + a**2)*sin(phi)*sin(theta), r_hor_p*cos(theta), 
                                   (phi, 1.0*pi, 1.5*pi), (theta, 0, pi)), 
                                  (sqrt(r_erg_p**2 + a**2)*cos(phi)*sin(theta), 
                                   sqrt(r_erg_p**2 + a**2)*sin(phi)*sin(theta), r_erg_p*cos(theta), 
                                   (phi, 0.0*pi, 0.5*pi), (theta, 0, pi)), 
                                  xlim=[-10.5, 10.5], ylim=[-10.5, 10.5], size=(10, 10), 
                                  xlabel="x", ylabel="y", zlabel="z",
                                  title="Kerr, M={}, a={}\n".format(M, a), show=False)
    p[0].surface_color = lambda x, y, z : z
    p[1].surface_color = lambda x, y, z : -z
    p[2].surface_color = lambda x, y, z : y
    p[3].surface_color = lambda x, y, z : x
    p.show()
```

```python
figura_Kerr_3D(4.2, 4.0)
```

```python
figura_Kerr_3D(4.2, 3.3)
```
