```python
from sympy import Symbol, sin, cos, sqrt, pi
from sympy.plotting.plot import plot_parametric as plotp
```

```python
theta = Symbol('theta')
```

```python
def r_erg_m(M, a, theta):
    return M / 2 - sqrt(M ** 2 / 4 - a ** 2 * cos(theta) ** 2 - Q ** 2)
```

```python
def r_erg_p(M, a, theta):
    return M / 2 + sqrt(M ** 2 / 4 - a ** 2 * cos(theta) ** 2 - Q ** 2)
```

```python
def r_hor_m(Q, M, a):
    return M / 2 - sqrt(M ** 2 / 4 - a ** 2 - Q ** 2)
```

```python
def r_hor_p(Q, M, a):
    return M / 2 + sqrt(M ** 2 / 4 - a ** 2 - Q ** 2)
```

```python
m = 8.2
a = 2
Q = 2
```

```python
inner_ergosurface = plotp(
    (
        r_erg_m(m, a, theta) * sin(theta),
        r_erg_m(m, a, theta) * cos(theta)
    ),
    (theta, 0, 2 * pi),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
)
```

```python
inner_event_horizon = plotp(
    (
        r_hor_m(Q, m, a) * sin(theta),
        r_hor_m(Q, m, a) * cos(theta)
    ),
    (theta, 0, 2 * pi),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
    )
```

```python
outer_event_horizon = plotp(
    (
        r_hor_p(Q, m, a) * sin(theta),
        r_hor_p(Q, m, a) * cos(theta)
    ),
    (theta, 0, 2 * pi),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
    )
```

```python
outer_ergosurface = plotp(
    (
        r_erg_p(m, a, theta) * sin(theta),
        r_erg_p(m, a, theta) * cos(theta)
    ),
    (theta, 0, 2 * pi),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
)
```

```python
black_hole = plotp(
    (
        r_erg_m(m, a, theta) * sin(theta),
        r_erg_m(m, a, theta) * cos(theta)
    ),
    (
        r_hor_m(Q, m, a) * sin(theta),
        r_hor_m(Q, m, a) * cos(theta)
    ),
    (
        r_hor_p(Q, m, a) * sin(theta),
        r_hor_p(Q, m, a) * cos(theta)
    ),
    (
        r_erg_p(m, a, theta) * sin(theta),
        r_erg_p(m, a, theta) * cos(theta)
    ),
    (theta, 0, 2 * pi), aspect_ratio=(1,1),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
)
```

```python
black_hole[0].line_color = 'blue'
black_hole[1].line_color = 'green'
black_hole[2].line_color = 'orange'
black_hole[3].line_color = 'red'
black_hole.show()
```
