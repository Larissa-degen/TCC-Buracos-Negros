```python
from sympy import *
from sympy.plotting.plot import plot_parametric
```

```python
def figura_Schwarzschild(M):
    theta = Symbol('theta')
    r_hor = 2*M
    plot_parametric((r_hor*sin(theta), r_hor*cos(theta)), (theta, 0, 2*pi), aspect_ratio=(1, 1), 
                    xlim=[-10, 10], ylim=[-10, 10], xlabel="x", ylabel="z", legend=True, label='hor', 
                    title="Schwarzschild, M={}\n".format(M), size=(8, 8))
```

```python
figura_Schwarzschild(4.2)
```

```python
figura_Schwarzschild(2.1)
```
