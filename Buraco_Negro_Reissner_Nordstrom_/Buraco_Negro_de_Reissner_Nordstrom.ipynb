```python
from sympy import Symbol, sin, cos, sqrt, pi
from sympy.plotting.plot import plot_parametric as plotp
```

```python
theta = Symbol('theta')
```

```python
def r_hor_m(Q, M):
    return 0.5 * (M - sqrt(M ** 2 - 4 * Q ** 2))
```

```python
def r_hor_p(Q, M):
    return 0.5 * (M + sqrt(M ** 2 - 4 * Q ** 2))
```

```python
m = 8.2
Q = 2
```

```python
inner_event_horizon = plotp(
    (
        r_hor_m(Q, m) * sin(theta),
        r_hor_m(Q, m) * cos(theta)
    ),
    (theta, 0, 2 * pi),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
    )
```

```python
outer_event_horizon = plotp(
    (
        r_hor_p(Q, m) * sin(theta),
        r_hor_p(Q, m) * cos(theta)
    ),
    (theta, 0, 2 * pi),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
    )
```

```python
black_hole = plotp(
    (
        r_hor_m(Q, m) * sin(theta),
        r_hor_m(Q, m) * cos(theta)
    ),
    (
        r_hor_p(Q, m) * sin(theta),
        r_hor_p(Q, m) * cos(theta)
    ),
    (theta, 0, 2 * pi), aspect_ratio=(1,1),
    xlim=[-15, 15], ylim=[-10, 10],
    xlabel="x", ylabel="z"
)
```

```python
black_hole[0].line_color = 'blue'
black_hole[1].line_color = 'red'

black_hole.show()
```
