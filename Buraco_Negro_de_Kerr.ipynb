```python
from sympy import *
from sympy.plotting.plot import plot_parametric
```

```python
def figura_Kerr(M, a):
    theta = Symbol('theta')
    r_erg_m = M - sqrt(M**2 - (a**2)*(cos(theta)**2))
    r_erg_p = M + sqrt(M**2 - (a**2)*(cos(theta)**2))
    r_hor_m = M - sqrt(M**2 - a**2)
    r_hor_p = M + sqrt(M**2 - a**2)
    p = plot_parametric((sqrt(r_erg_m**2 + a**2)*sin(theta), r_erg_m*cos(theta)), 
                        (sqrt(r_hor_m**2 + a**2)*sin(theta), r_hor_m*cos(theta)),
                        (sqrt(r_hor_p**2 + a**2)*sin(theta), r_hor_p*cos(theta)),
                        (sqrt(r_erg_p**2 + a**2)*sin(theta), r_erg_p*cos(theta)), 
                        (theta, 0, 2*pi), aspect_ratio=(1, 1), xlim=[-10.5, 10.5], ylim=[-7.5, 7.5], 
                        xlabel="x", ylabel="z", legend=True, 
                        title="Kerr, M={}, a={}\n".format(M, a), size=(8, 8), show=False)
    p[0].line_color = 'orange'
    p[0].label = 'erg-'
    p[1].line_color = 'red'
    p[1].label = 'hor-'
    p[2].line_color = 'blue'
    p[2].label = 'hor+'
    p[3].line_color = 'green'
    p[3].label = 'erg+'
    p.show()
```

```python
figura_Kerr(4.2, 4.0)
```

```python
figura_Kerr(4.2, 3.3)
```
